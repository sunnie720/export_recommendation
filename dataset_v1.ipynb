#### population 피벗테이블 형식으로 되어있어서 스프레드시트 형태로 풀어주기

```python
population.columns
```

```python
# 연도별로 총인구, 인구증가율 데이터만 분리
pop2019 = population[['국가별(1)', '국가별(2)', '2019', '2019.1']]
pop2020 = population[['국가별(1)', '국가별(2)', '2020', '2020.1']]
pop2021 = population[['국가별(1)', '국가별(2)', '2021', '2021.1']]
pop2019.head()
```

```python
# 연도행 추가하고 컬럼명 변경 및 불필요한 문자열 제거
def unpivot_population(df):
    df['year'] = df.columns[2]
    df.columns = ['권역', '국가', '총인구', '인구증가율', 'year']
    df.drop(0, inplace=True)
    df.총인구 = df.총인구.apply(lambda x: x.split('.')[0])
    df.인구증가율 = df.인구증가율.apply(lambda x: x.split('(')[0][:-1]) # 1.2% -> 0.012
    df = df[['year', '국가', '총인구', '인구증가율']]
    return df

pop2019 = unpivot_population(pop2019)
pop2020 = unpivot_population(pop2020)
pop2021 = unpivot_population(pop2021)
pop2019.head()
```

```python
# 연도별 데이터 합치기
pop = pd.concat([pop2019, pop2020, pop2021])

# '소계' 행 제거
pop = pop[pop.국가 != '소계']
print('국가 수:', pop.국가.nunique())
display(pop) 
```

```python
pop.isna().sum()
```

#### GDP 피벗테이블 형식으로 되어있어서 스프레드시트 형태로 풀어주기

```python
display(GDP.head())
print(GDP.columns)
```

```python
# GDP 성장률 - 2013년 이후 데이터만 추출
GDP_growth = pd.DataFrame()
year_list = [i for i in range(2013, 2023)]
for year in year_list:
    GDP_temp = GDP[['Country Name', str(year), str(year-1)]]
    GDP_temp['year'] = year
    GDP_temp['GDP_growth'] = (GDP_temp[str(year)] - GDP_temp[str(year-1)]) / GDP_temp[str(year-1)] 
    GDP_temp = GDP_temp[['Country Name', 'year', 'GDP_growth']]
    GDP_temp.rename(columns={'Country Name':'country'}, inplace=True)
    GDP_growth = pd.concat([GDP_growth, GDP_temp])
    
print('국가 수:', GDP_growth.country.nunique())
display(GDP_growth) 
```

```python
# 결측치 개수 확인
GDP_growth.isna().sum()
```

```python
# 결측치 확인
GDP_growth[GDP_growth.GDP_growth.isna()]
```

```python
# 결측치들은 원본 데이터에서도 결측치이므로 제거
display(GDP[GDP['Country Name']=="Channel Islands"])
display(GDP[GDP['Country Name']=="Eritrea"])
display(GDP[GDP['Country Name']=="Gibraltar"])
```

```python
# 결측치 제거
GDP_growth.dropna(inplace=True)

display(GDP_growth.isna().sum())
print('데이터 수:', GDP_growth.shape[0])
print('국가 수:', GDP_growth.country.nunique())
```

```python
# rfmd 데이터와 맞출 수 있게 '품목번호' 생성
tariff['품목코드'] = tariff['품목번호'].apply(lambda x: x[:2])
tariff
```

```python
print(rfmd.품목코드.nunique())
print(tariff.품목코드.nunique())
```

```python
print(set(tariff.품목코드.unique()) - set(rfmd.품목코드.unique())) # rfmd에는 없고 tarrif에만 있는 품목코드
print(set(rfmd.품목코드.unique()) - set(tariff.품목코드.unique())) # tarrif에는 없고 rfmd에만 있는 품목코드
```

```python
# tariff 테이블에서 품목코드별 
item_tariff = tariff.groupby('품목코드')['관세율'].agg(['min', 'max', 'mean', 'median']).reset_index()
item_tariff
```

#### rfmd + tariff 병합

```python
rfmd_tariff = rfmd.copy()
rfmd_tariff = pd.merge(rfmd_tariff, item_tariff, on='품목코드', how='left')

print(rfmd_tariff.shape)
display(rfmd_tariff.head())
```
